```python
import os
import urllib.request
import ssl
from torch_geometric.datasets import Planetoid
import torch_geometric.transforms as T

# 1. Temporarily bypass Mac's standard SSL verification for manual downloads
ssl._create_default_https_context = ssl._create_unverified_context

# 2. Define the exact path (let's keep it inside your project folder)
data_dir = './data/Cora'
raw_dir = os.path.join(data_dir, 'Cora', 'raw') # PyG creates a nested 'Cora/raw' folder
os.makedirs(raw_dir, exist_ok=True)

# 3. The exact files PyG needs to skip the download step
files = [
    'ind.cora.x', 'ind.cora.y', 'ind.cora.tx', 'ind.cora.ty',
    'ind.cora.allx', 'ind.cora.ally', 'ind.cora.graph', 'ind.cora.test.index'
]
base_url = "https://github.com/kimiyoung/planetoid/raw/master/data/"

print("Downloading Cora files manually...")
for f in files:
    url = base_url + f
    save_path = os.path.join(raw_dir, f)
    if not os.path.exists(save_path):
        print(f"Downloading {f}...")
        urllib.request.urlretrieve(url, save_path)
    else:
        print(f"{f} already exists locally.")

print("\nLoading Cora into PyTorch Geometric...")
# 4. Initialize PyG pointing to the exact same root directory
dataset = Planetoid(root=data_dir, name='Cora', transform=T.NormalizeFeatures())
data = dataset[0]

print("✅ Dataset loaded successfully!")
```
